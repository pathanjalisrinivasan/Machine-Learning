```python
#Treating outliers
cont = data_df.dtypes[(data_df.dtypes!='uint8')&(data_df.dtypes!='object')].index
def remove_outlier(col):
    sorted(col)
    Q1,Q3= np.percentile(col,[25,75])
    IQR= Q3-Q1
    lower_range= Q1-(1.5*IQR)
    upper_range= Q3+(1.5*IQR)
    return lower_range, upper_range
```

```python
for column in data_df[cont].columns:
    lr,ur= remove_outlier(data_df[column])
    data_df[column]= np.where(data_df[column]>ur,ur,data_df[column])
    data_df[column]= np.where(data_df[column]<lr,lr,data_df[column])
```

```python
# empty list that will hold accuracy scores
ac_scores = []

# perform accuracy metrics for values from 1,3,5....19
for k in range(1,20,2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_trains, y_trains)
    # evaluate test accuracy
    scores = knn.score(X_tests, y_tests)
    ac_scores.append(scores)

# changing to misclassification error
MCE = [1 - x for x in ac_scores]
MCE
```

```python
#Plot misclassification error vs k (with k value on X-axis)
plt.plot(range(1,20,2), MCE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```

2.3 Which word occurs the most number of times in his inaugural address for each president? Mention the top three words. (after removing the stopwords)

```python
tokenizer = RegexpTokenizer(r'\w+')
roosevelt_no_punc = tokenizer.tokenize(Roosevelt)
set(w.title() for w in roosevelt_no_punc if w.lower() not in stopwords.words())
word_count_dict1 = Counter(w.title() for w in roosevelt_no_punc if w.lower() not in stopwords.words())

print("Top three words in Roosevelt's speech(after removing the stopwords):",word_count_dict1.most_common(3))
```

```python
tokenizer = RegexpTokenizer(r'\w+')
kennedy_no_punc = tokenizer.tokenize(Kennedy)
set(w.title() for w in kennedy_no_punc if w.lower() not in stopwords.words())
word_count_dict2 = Counter(w.title() for w in kennedy_no_punc if w.lower() not in stopwords.words())

print("Top three words in Kennedy's speech(after removing the stopwords):",word_count_dict2.most_common(3))
```

```python
tokenizer = RegexpTokenizer(r'\w+')
nixon_no_punc = tokenizer.tokenize(Nixon)
set(w.title() for w in nixon_no_punc if w.lower() not in stopwords.words())
word_count_dict3 = Counter(w.title() for w in nixon_no_punc if w.lower() not in stopwords.words())

print("Top three words in Kennedy's speech(after removing the stopwords):",word_count_dict3.most_common(3))
```
